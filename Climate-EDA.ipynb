# Exploratory Data Analysis of climate change

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (12,8)
```

```python
data = pd.read_csv('climate-cleaned.csv')
data.head()
```

```python
data.describe(include='all')
```

```python
colToPlot = list(data.columns[3:9])
colToPlot
```

```python
fig,ax = plt.subplots(3,2,figsize=(20,14))
plt.suptitle('Trend over years (2000-2022)')

for i,ax in enumerate(fig.axes):
    sns.lineplot(x='year', y=colToPlot[i],data=data, hue='Continent',style='Continent',ax=ax,ci=None)
    ax.set(title='Average '+ax.get_ylabel().lower()+' change over years')
    
fig.tight_layout()
```

```python
fig,ax = plt.subplots(3,2,figsize=(20,14))
plt.suptitle('Trend by months')

for i,ax in enumerate(fig.axes):
    sns.lineplot(x='monthName', y=colToPlot[i],data=data, hue='Continent',style='Continent',ax=ax,ci=None)
    ax.set(title='Average '+ax.get_ylabel().lower()+' by month')
    
fig.tight_layout()
```
