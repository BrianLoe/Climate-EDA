# Exploratory Data Analysis of climate change

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (12,8)
```

```python
data = pd.read_csv('climate-cleaned.csv')
data.head()
```

```python
data.describe(include='all')
```

```python
fig,ax = plt.subplots(3,2,figsize=(20,14))
plt.suptitle('Trend over years (2000-2022)')
sns.lineplot(x='year', y='Temperature',data=data, hue='Continent',style='Continent',ax=ax[0,0],ci=None)
sns.lineplot(x='year', y='CO2 Emissions',data=data, hue='Continent',style='Continent',ax=ax[0,1],ci=None)
sns.lineplot(x='year', y='Sea Level Rise',data=data, hue='Continent',style='Continent',ax=ax[1,0],ci=None)
sns.lineplot(x='year', y='Precipitation',data=data, hue='Continent',style='Continent',ax=ax[1,1],ci=None)
sns.lineplot(x='year', y='Humidity',data=data, hue='Continent',style='Continent',ax=ax[2,0],ci=None)
sns.lineplot(x='year', y='Wind Speed',data=data, hue='Continent',style='Continent',ax=ax[2,1],ci=None)
for i,ax in enumerate(fig.axes):
    ax.set(title=ax.get_ylabel()+' change over years')
fig.tight_layout()
```
